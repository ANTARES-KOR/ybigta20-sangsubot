```python
from chat_downloader import ChatDownloader

# 나는 왜 테이저건을 맞았는가 EP.1
url = 'https://www.youtube.com/watch?v=D-7HWnXmeOk'
chat = ChatDownloader().get_chat(url)       # create a generator
chats = []

for message in chat:                        # iterate over messages
    chats.append([message['message'], message['author']['name'], message['time_text']])
```

```python
import pandas as pd

df = pd.DataFrame(chats, columns=[['댓글', '작성자', '작성시간']])
```
